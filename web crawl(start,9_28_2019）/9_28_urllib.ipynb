```python
# urlopen
# urlopen 只能打开网页相关信息，不能做高级的传送信息到网页上
#urllib.request.urlopen(url, data=None, [timeout, ]*, cafile=None, capath=None, cadefault=False, context=None)
import  urllib.request

url = "https://www.sec.gov/Archives/edgar/data/1621563/000119312515080375/d789569ds1a.htm"
# 打开url源代码
response = urllib.request.urlopen(url)
# 读取
print(response.read().decode('utf-8'))
```

```python
# post 类型请求
import urllib.parse
import urllib.request

url ="http://httpbin.org/post"
# 输入data，bytes类型，将英文encode
post_data = bytes(urllib.parse.urlencode({'world':'hello'}), encoding = 'utf8')
response = urllib.request.urlopen(url,data = post_data)
print(response.read())
```

```python
#在规定时间内返回请求
import urllib.request

url ="http://httpbin.org/get"
response = urllib.request.urlopen(url,timeout = 1)
print(response.read())
```

```python
#没有返回的话返回错误提示
import socket
import urllib.request
import urllib.error

url ="http://httpbin.org/get"
# try: lets you test a block of code for errors
# except: lets you handle the error
try:
    response = urllib.request.urlopen(url,timeout = 0.1)
except urllib.error.URLError as e:
    if isinstance(e.reason,socket.timeout):
        print("TimeOut")
```

```python
# response 类型
import urllib.request

url = "http://httpbin.org"
response = urllib.request.urlopen(url)
print(type(response))
```

```python
# 状态码，响应头
import urllib.request

url = "http://httpbin.org"
response = urllib.request.urlopen(url)
# 状态码
print(response.status)
# 响应头
print(response.getheaders())
# 响应头里其中一项-- server
print(response.getheader("Server"))
```

```python
# Request 可以传送更多信息到网页上
# 读取页面信息
import urllib.request

url = "http://httpbin.org"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
print(response.read().decode('utf-8'))
```

```python
# 更加方便地指定请求的解决方式，可以加一些参数，额外的数据
from urllib import request,parse
url = "http://httpbin.org/post"
headers = {
    "User-Agent":'Mozilla',
    "Host": "httpbin.org"
}
dict = {
    "name":"Germey"
}
post_data = bytes(parse.urlencode(dict),encoding = 'utf8')
req = request.Request(url = url,data = post_data,headers =headers,method = "POST")
response = request.urlopen(req)
print(response.read().decode('utf-8'))
```

```python
from urllib import request,parse

url = "http://httpbin.org/post"
dict = {
    "name":"Germey"
}
post_data = bytes(parse.urlencode(dict),encoding = 'utf8')
req = request.Request(url = url,data = post_data,method = "POST")
req.add_header("User-Agent","Mozilla")
response = request.urlopen(req)
print(response.read().decode('utf-8'))
```
